```python
from pptree import *

class TreeNode:
    def __init__(self, key):
        self.key = key
        self.children = []
        self.degree = 0
    
    def add_child(self, tree):
        
        if tree.key > self.key: 
        
            self.children.append(tree) 

            # this is because the degree will make 
            self.degree = len(self.children) + tree.degree
            
            self.children.sort(key=lambda x: x.key)
            
    def search_tree(self, key):
        
        if len(self.children) == None:
            return None
        
        
        for child in self.children:
            if child.key == key:
                return [self, child]
            else:
                child.search_tree(key)
    
    def add_nodes_for_print(self, parent):
        

        

        if len(self.children) > 0:
            for child in self.children:

                k = Node(str(child.key), parent)

                child.add_nodes_for_print(k)
        
        

        
        
```
