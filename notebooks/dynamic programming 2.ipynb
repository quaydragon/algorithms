
2. Jars on a ladder problem. Given a ladder of n rungs and k identical glass jars, one has to design an experiment of dropping jars from certain rungs, in order to find the highest rung (HS) on the ladder from which a jar doesn’t break if dropped.

Idea: With only one jar (k=1), we can’t risk breaking the jar without getting an answer. So we start from the lowest rung on the ladder, and move up. When the jar breaks, the previous rung is the answer; if we are unlucky, we have to do all n rungs, thus n trials. Now lets think of k=log(n): with log(n) or more jars, we have enough jars to do binary search, even if jars are broken at every rung. So in this case we need log(n) trials. Note that we can’t do binary search with less than log(n) jars, as we risk breaking all jars before arriving at an answer in the worst case.
Your task is to calculate q = MinT(n,k)= the minimum number of dropping trials any such experiment has to make, to solve the problem even in the worst/unluckiest case (i.e., not running out of jars to drop before arriving at an answer). MinT stands for Minumum number of Trials.

### (a) Explaining the Optimal Solution

The optimal solution to this subproblem is when we identify the highest rung that a jar can be dropped from and does not break using the least amount of drops. 

OPTSOL = First Rung Jar Breaks - 1

In setting up the problem we get:

n = the rungs 
k = the jars

C[n, k] = The highest rung the jar can be dropped from in the least amount of drops. MinT(n, k). We want to minimize k while maximizing n prior to breaking the jar. 



Now to view this recursively. The minimum outside of the bracket is to state the minimum iterations it takes to find the optimal solution. The way the program is designed will try and find the minimum of these actions. 



```python
%%latex
l = height of ladder section being explored

$C[n, k] =  \min \begin{cases}
       0 & \text{if n = 0 or k = 0}\\
      1 + C[n + 1, k] & \text{if k = 1  and n is not the highest rung + 1. n must start at the lowest past break point or past climb point.}\\
      1 + C[n + 1, k - 1] & \text{if k = 1  and n is the highest rung + 1. n must start at the lowest past break point or past climb point.}\\
      1 +  \max (C[l / 2, k]) \text{ explore up} & \text{If k > 1 and n is not the highest rung + 1. n starts at half the given ladder size of section being explored.}\\
      1 + \max (C[l / 2, k]) \text{ explore down} & \text{If k > 1 and n is the higest rung. n starts at half the given ladder size of section being explored.}\\


    \end{cases}$  
```
