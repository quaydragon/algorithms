11. (10 pts) Exercise 12.2-5.



The successor should be the most left child on the right side of the tree given a root node and the predecessor should be the most right child on the left side of a tree given a root node. This means it is impossible for a sucessor to have a left child because this means it would be closer in value to the root node. The same notion can be applied to the predecessor, if the predecessor had a right child the value of that child would be closer to the root node therefore making the child the predessor. Therefore, it is impossible for the successor to have a left child and impossible for the predecessor to have a right child.
