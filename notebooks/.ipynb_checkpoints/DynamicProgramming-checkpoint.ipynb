```python
# Sub Sequence Problem 


# what if you save the count for every single number 

#table in n time complete 

sequence = [1,3,2,1,5,6,2,8]

count_table = [0]*len(sequence)
lis = [[]] * len(sequence)
currentcount = 0




for n in range(len(sequence)):
    lis[n] = [sequence[n]]
    count_table[n] += 1
    for j in range(n, len(sequence)):
        if sequence[j] > lis[n][-1]:
            lis[n] = lis[n] + [sequence[j]]
            count_table[n] +=1
    

def getMaxLength(table):
    return max(table)
            
            

def getMaxSequence(table, lis):
    
    max_index = 0
    maximum = 0
    for i in range(len(table)):
        if table[i] > maximum:
            max_index = i
            maximum = table[i]
    return lis[max_index]
    
    
print(getMaxLength(count_table))
print(getMaxSequence(count_table, lis))
        
```

```python
# 15 - 5

#first build out the initial functions

def copy(x, z, i, j, moves):
#     if x[i] == y[j]:
#         print('returning copy')
    z[j] = x[i]
    i += 1
    j += 1
    moves = moves
    moves.append('copy')
#         return (x, z, i, j, moves)
    print("entered copy but did not copy")
    return(x, z, i, j, moves)

def replace(x, z, i, j, c, moves):
    print('returning replace')
    z[j] = c
    i += 1
    j += 1
    moves = moves
    moves.append(f'replace {z[j - 1]}')
    return(x, z, i, j, moves)

def delete(x, z, i, j, moves):
    print('returning delete')
    i += 1
    moves = moves
    moves.append('delete')
    return(x, z, i, j, moves)

def insert(x, z, i, j, c, moves):
    print(f'returning insert {z[j]}')
    print(moves)
    z[j]  = c
    j += 1
    moves = moves
    moves.append(f'insert {z[j - 1]}')
    return(x, z, i, j, moves)

def twiddle(x, z, i, j, moves):
    print('returning twiddle')
    z[j] = x[i + 1]
    z[j + 1] = x
    i = i + 2
    j = j + 2
    
    return (x, z, i, j, moves)


def kill(x, z, i, j):
    print('returning kill')
    i = len(x) + 1
    return (x, z, i, j)
    
    
        
    
    
    
```
