```python

class Node:
    def __init__(self, keyval=None, dataval=None):
        self.keyval = keyval
        self.dataval = dataval
        self.nextval = None

class SLinkedList:
    def __init__(self):
        self.headval = None
            

    def listprint(self, file):
        printval = self.headval
        while printval is not None:
            print (printval.dataval, printval.keyval)
            file.write(f'{printval.dataval,} {printval.keyval} \n')
            printval = printval.nextval




## hash function

class HashClass:
    def __init__(self, size):     
        self.size = size
        self.buckets = [None] * self.size

    def hashit(self,key):
        # loook at rolling hash
        
        return len(key) % self.size


        
    
    def insert_key(self,key):
        hashed = self.hashit(key)

        if self.buckets[hashed] == None:
            list1 = SLinkedList()
            list1.headval = Node(key, 1)
            
            #what we should be doing is immediately creating a linked list
            self.buckets[hashed] = list1
        
        else:
            list2 = self.buckets[hashed]
            
            # need to check if the key value is in the linked list
            
            key_bool = False
            lastnode = False
            node = list2.headval
            
            # add value to key that already exists
            while True:                
                if node.keyval == key:
                    
                    ## update the count
                    node.dataval += 1
                    key = True
                    break    
                elif node.nextval == None and node.keyval != key:
                    node.nextval = Node(key, 1)
                    break
                node = node.nextval

                

                    
                
                
                


        
    
    def delete_key(self, key):
        hashed = self.hashit(key)
        linked = self.buckets[hashed]
        
        node_prev = None
        node = linked.headval
        while True:        
            if node.keyval == key:
                if node_prev == None:
                    self.buckets[hashed] = node.nextval
                else:
                    node_prev.nextval = node.nextval
                break      

            node_prev = node    
            node = node.nextval
        
        
        linked.listprint()
            
        
    
    
    def find(self, key):
        hashed = self.hashit(key)
        
        linked = self.buckets[hashed]
        
        if linked:
            node = linked.headval
            while True:
                if node.keyval == key:
                    return node
                
                node = node.nextval
                if node == None:
                    break
                    
    def list_all(self):
        
        file = open("write.txt","w")
        file.truncate(0)
        for bucket in self.buckets:
            
            if bucket != None:
                bucket.listprint(file)
    
```
